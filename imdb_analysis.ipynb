Convert Box Office Data to Int to Enable Analysis

*e*.g. Currently exists as $10.00M, outputs to 10000000




```python
# Convert Box Office to Float64 Using Regex to enable graphical analysis
# Regex 1 (First half of Decimal) = (\d{1,3})\
# Regex 2 (Second half of Decimal)= (\d{2})

import re
import numpy as np 

df['Gross'] = df['Gross'].apply(str)

#regex functions to obtain values
def first_reg(Gross):
  x = re.search('(\d{1,3})', Gross)
  return(x.group())

def second_reg(Gross):
  y = re.search('(\d{2})', Gross)
  try:
      return(y.group(1))
  finally:
      return(y.group(0))
  
df['Million 1'] = df['Gross'].apply(first_reg)
df['Million 2'] = df['Gross'].apply(second_reg)

# Convert Values to get float for Box Office Value
df['Box Office Value'] = df['Million 1']+df['Million 2']+'0000'
df['Box Office Value'] = df['Box Office Value'].astype(int)

# Remove Extra Columns and Check the Data
df=df.drop(["Million 1"], axis = 1)
df=df.drop(["Million 2"], axis = 1)
df.head()
```

Sentimental Analysis Script

Iterate through each movie reference by sub-group and establish the polarity and subjectivity of the reference. 

Store data in Sent Analysis DF.

```python
# Import Libraries for Sentimental Analysis 
from bs4 import BeautifulSoup
from textblob import TextBlob

## List of Diseases Terms and Sub-Groups to Search for
general = ["ILLNESS", "AFFLICTION", "AILMENT", "DISORDER", "INFECTION", "INFIRMITY", "CONTAMINATION", "DISORDER", "EPIDEMIC", "FEVER", "FLU ", "ILLNESS", "INFECTION", "INFLAMMATION", "MALADY", "PLAGUE", "SICKNESS", "STROKE", "SYNDROME", "VIRUS", " ACHE ", "CONTAGION", "CONVULSION", "DEBILITY", "DECREPITUDE", "ENDEMIC", "HEMORRHAGE", "INFIRMITY", "SEIZURE", "UNHEALTHY", "ILL HEALTH", "PATHOSIS", "SICKLINESS"]
neuro = ["EPILEPSY", "STROKE", "AANEURYSM", "PARKINSON'S", "BIPOLAR", "DEMENTIA", "ALZHEIMER’S", "DEPRESSION", "SCHIZOPHRENIA"]
respiratory = ["RHINITIS", "ASTHMA", "CYSTIC FIBROSIS", "ADENOVIRUS", "MERS-COV", "PNEUMONIA", "BRONCHITIS", "CHRONIC OBSTRUCTIVE PULMONARY DISEASE", "COUGH", "EMPHYSEMA", "TUBERCULOSIS"]
eye = ["BLEPHARITIS", "CATARACT", "DRY EYES", "GLAUCOMA", "RETINAL DETACH", "STRABISMUS", "STYE", "CHALAZION", "COLOUR BLINDNESS", "HYPEROPIA", "MYOPIA", "EYE TWITCH"]
endocrine = ["HYPERTHYROIDISM", "HYPOTHYROIDISM", "INSOMNIA", "THYROID DISEASE", "DIABETES"]
muscle = ["PARALYSIS", "SCIATICA", "VARICOSE", "BELL'S PALSY"]
blood = ["ANAEMIA", "HYPERTENSION", "THALASSEMIA", "ANEURYSM", "THROMBOSIS"]
boneandjoint = ["ARTHRITIS", "BACK PAIN", "BRITTLE BONE", "OSTEOGENESIS IMPERFECTA", "CARPAL TUNNEL", "OSTEOMYELITITS", "OSTEOPOROSIS", "SLIPPED DISC", "CERVICAL SPONDYLOSIS"]
skin = ["ALLERGY", "ANAPHYLAXIS", "ATOPIC DERMATITIS", "DERMATITIS", "SEBORRHEIC DERMATITIS", "BURNS", "CELLULITIS", "CONTACT DERMATITIS", "ECZEMA", "FOLLICULITIS", "KELOIDS"]
oral = ["CLEFT LIP", "ULCERS", "ORAL THRUSH", "GUM DISEASE", "PERIODONTITIS", "TEETH SENSITIVITY", "DRY MOUTH"]
cancer = ["CANCER", "ORAL CANCER", "BREAST CANCER", "THYROID CANCER", "CANCERS IN CHILDREN", "LUNG CANCER", "PROSTATE CANCER", "CERVICAL CANCER", "OVARIAN CANCER", "SKIN CANCER", "BOWEL CANCER", "COLORECTAL CANCER", "OESOPHAGEAL CANCER", "UTERINE CANCER", "CANCER"]
other = ["SWINE FLU", " UTI ", "URINE TRACT INFECTION", "EBOLA", "DENGUE", "VERTIGO", "BIRD FLU", "CYSTS", "OVARIAN CYSTS", "MALARIA", " HIV ", " AIDS ", "HIV/AIDS" "HERNIA", "ENDOMETRIOSIS", "CHIKUNGUNYA", "RABIES", "POLYCYSTIC OVARY SYNDROME", "BROKEN BONE"]
autoimmune = ["LUPUS", "GRAVES DISEASE", "LICHEN PLANUS", "PSORIASIS", "VITILIGO", "ADDISON’S", "CELIAC", "CROHN’S", "ULCERATIVE COLITIS", "SJOGERN’S SYNDROME", "ANAEMIA", "ANAEMIA", "ANKYLOSING SPONDYLITIS", "LYME DISEASE", "SARCOIDOSIS", "RHEUMATOID ARTHRITIS", "FIBROMYALGIA", "MYASTHENIA GRAVIS", "MULTIPLE SCLEROSIS", "OPTIC NEURITIS", "RESTLESS LEG SYNDROME", "MENIERE’S DISEASE", "KAWASAKI DISEASE"]
biotech = ["BIOTECHNOLOGY", "GENE-SPLICING", "GENE RESEARCH", "DNA TECHNOLOGY", "GENETICALLY MODIFIED", "HUMAN CLONING", "BIOENGINEERING", "ARTIFICIAL INTELLIGENCE", "GENETIC ENGINEERING", "BIONIC EYE", "BIONIC LEG", "BIONIC EAR", "BIONIC LIMB", "BIONIC ARM", " TISSUE ENGINEERING", " 3D PRINTED ORGANS", "EXOSUITS", "EXO-SUITS", "MECH SUITS", "MECH-SUITS"]

# List Terms as List
terms = [general,neuro,respiratory,eye,endocrine,muscle,blood,boneandjoint,skin,oral,cancer,autoimmune,other,biotech]

#Functions to enable Sentimental Analysis on surrounding text
subj_threshold = 150
def getSubjectivity(input):
  output = TextBlob(input).sentiment.subjectivity
  return output
def getPolarity(input):
  output = TextBlob(input).sentiment.polarity
  return output

# Search each movie synopsis for any reference to disease terms and count for each subgroup
# Appends these counts to the dataframe for further analysis

# Search for General
for movie in synopsis_data:
    movie_syn = synopsis_data[movie]['Synopsis']
    movie_syn = BeautifulSoup(movie_syn, "lxml").text
    movie_syn = movie_syn.upper()
    i = 0
    # Iterate through each disease type
    for group in terms:
        i=i+1 
        group_dict = {1:"general", 2:"neuro", 3:"respiratory", 4: "eye", 5: "endocrine", 6:"muscle", 7: "blood", 8: "boneandjoint", 9: "skin", 10: "oral", 11: "cancer", 12: "other", 13: "autoimmune", 14: "biotech"}
        subgroup = group_dict[int(i)]

        for disease in group:
            if movie_syn.count(disease) > 0:
                  # Find surrounding text for Sentimental Analysis        
                  disease_loc = movie_syn.find(disease)
                  dist_to_end = len(movie_syn)-disease_loc
                  #find start location for surrounding text
                  if disease_loc < subj_threshold:
                      start_loc = 0
                  else:
                      start_loc = disease_loc-subj_threshold
                  # Find end location for surrounding text
                  if dist_to_end < subj_threshold:
                      end_loc = len(movie_syn)
                  else:
                      end_loc = disease_loc+subj_threshold

                  surrounding_text = movie_syn[start_loc:end_loc]
                  output_subj = getSubjectivity(surrounding_text)
                  output_pol = getPolarity(surrounding_text)

                  new_row = [movie, subgroup, surrounding_text, output_subj, output_pol]
           
                  sent_df.loc[len(sent_df)]= new_row
                      
```
